```python
from psx import stocks, tickers
import datetime
```

```python
tickers = tickers()
today = datetime.date.today()
yesterday = yesterday = today - datetime.timedelta(days=1)
yesterday
```

```python
data = """BOP"""
stock_names = data.split()
data = stocks(stock_names, start=datetime.date(2023, 1, 1), end=yesterday)
```

```python
data
```

```python
df = data.reset_index()
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']
df.to_csv('actual.csv', index=False)
df.head()
```

```python
import pandas as pd
df = pd.read_csv("actual.csv")
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Calculate Simple Moving Average (SMA) over 3 days
df['SMA_3'] = df['Close'].rolling(window=3).mean()

# Calculate Exponential Moving Average (EMA) over 3 days
df['EMA_3'] = df['Close'].ewm(span=3, adjust=False).mean()

# Calculate Daily Returns
df['Daily_Return'] = df['Close'].pct_change()

# Calculate Rolling Standard Deviation over 3 days
df['Rolling_Std_3'] = df['Close'].rolling(window=3).std()

# Handling NaN values
df['SMA_3'].fillna(df['Close'], inplace=True)
df['EMA_3'].fillna(df['Close'], inplace=True)
df['Daily_Return'].fillna(0, inplace=True)
df['Rolling_Std_3'].fillna(df['Rolling_Std_3'].mean(), inplace=True)  # Optionally fill with the mean

# Save the modified DataFrame to a new CSV file
df.to_csv('modified.csv', index=False)
```
